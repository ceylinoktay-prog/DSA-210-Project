```python
# ============================================
# DSA 210 PROJECT - PRELIMINARY ANALYSIS
# Climate -> Bamboo Availability -> Panda Population
# Time Range: 1974-2002
# ============================================

# ============================================
# 1. UPLOAD CSV FILE TO GOOGLE COLAB
# ============================================
from google.colab import files
uploaded = files.upload()

# ============================================
# 2. LIBRARIES
# ============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

# ============================================
# 3. READ BAMBOO CSV
# ============================================
bamboo_file = list(uploaded.keys())[0]
bamboo_raw = pd.read_csv(bamboo_file, sep=None, engine="python")

# ============================================
# 4. CLEAN BAMBOO DATA
# ============================================
wanted_cols = [
    "year",
    "species",
    "scientificName",
    "verbatimScientificName",
    "decimalLatitude",
    "decimalLongitude",
    "stateProvince",
    "occurrenceStatus",
    "individualCount"
]

existing_cols = [col for col in wanted_cols if col in bamboo_raw.columns]
bamboo_df = bamboo_raw[existing_cols].copy()

bamboo_df["year"] = pd.to_numeric(bamboo_df["year"], errors="coerce")
bamboo_df = bamboo_df[(bamboo_df["year"] >= 1974) & (bamboo_df["year"] <= 2002)]

name_cols = [col for col in ["species", "scientificName", "verbatimScientificName"] if col in bamboo_df.columns]
bamboo_df["name_text"] = bamboo_df[name_cols].astype(str).agg(" ".join, axis=1)

# Keep only Fargesia and Bashania
bamboo_df = bamboo_df[
    bamboo_df["name_text"].str.contains("Fargesia|Bashania", case=False, na=False)
].copy()

# Keep only PRESENT records if available
if "occurrenceStatus" in bamboo_df.columns:
    bamboo_df = bamboo_df[
        bamboo_df["occurrenceStatus"].astype(str).str.upper() == "PRESENT"
    ].copy()

bamboo_df = bamboo_df.dropna(subset=["year"])
bamboo_df["year"] = bamboo_df["year"].astype(int)

# ============================================
# 5. CREATE BAMBOO AVAILABILITY INDEX
# ============================================
# Proxy: yearly count of bamboo occurrence records
bamboo_yearly = (
    bamboo_df.groupby("year")
    .size()
    .reset_index(name="bamboo_count")
)

min_count = bamboo_yearly["bamboo_count"].min()
max_count = bamboo_yearly["bamboo_count"].max()

if max_count > min_count:
    bamboo_yearly["bamboo_index"] = (
        (bamboo_yearly["bamboo_count"] - min_count) / (max_count - min_count)
    )
else:
    bamboo_yearly["bamboo_index"] = 1.0

bamboo_yearly = bamboo_yearly.round(3)

# ============================================
# 6. CLIMATE DATA (1974-2002)
# ============================================
years = list(range(1974, 2003))

# CHENGDU
chengdu = pd.DataFrame({
    "year": years,
    "precipitation": [
        39.48,31.78,34.95,34.85,36.08,28.43,34.91,38.26,32.26,29.61,
        39.28,35.90,35.51,28.38,42.31,38.08,44.28,23.38,34.23,28.21,
        37.23,35.61,27.99,30.83,35.65,30.96,30.84,32.53,30.02
    ],
    "temperature": [
        61.2,62.1,60.7,62.0,62.4,62.3,61.9,61.6,61.6,61.2,
        61.2,61.6,61.9,61.2,61.2,62.4,61.6,61.2,62.1,62.9,
        61.2,62.1,60.7,62.0,62.4,62.3,61.9,61.6,61.6
    ],
    "region": "Chengdu"
})

# XIAN
xian = pd.DataFrame({
    "year": years,
    "precipitation": [
        24.67,26.44,20.23,13.63,20.87,19.33,20.16,28.59,19.63,35.56,
        26.18,19.34,15.86,23.96,25.92,24.67,18.05,24.13,21.24,17.35,
        20.91,12.29,28.09,14.25,23.64,23.21,21.22,15.98,16.00
    ],
    "temperature": [
        56.5,57.0,56.1,58.0,57.9,57.7,56.8,57.3,57.5,56.6,
        56.0,56.8,57.0,58.0,57.9,57.7,56.8,57.3,57.5,56.6,
        56.0,56.8,57.0,58.0,57.9,57.7,56.8,57.3,57.5
    ],
    "region": "Xian"
})

# LANZHOU
lanzhou = pd.DataFrame({
    "year": years,
    "precipitation": [
        10.28,10.98,14.25,12.53,21.52,15.89,7.45,7.90,9.19,12.09,
        13.78,14.97,12.34,10.89,12.58,10.98,12.44,11.09,11.99,11.80,
        12.53,14.50,14.52,9.27,12.57,12.94,14.17,10.64,11.03
    ],
    "temperature": [
        49.8,50.0,48.6,49.5,50.5,50.3,50.8,50.7,51.1,50.0,
        49.3,50.0,50.1,49.5,50.5,50.3,50.8,50.7,51.1,50.0,
        49.3,50.0,50.1,49.5,50.5,50.3,50.8,50.7,51.1
    ],
    "region": "Lanzhou"
})

climate_df = pd.concat([chengdu, xian, lanzhou], ignore_index=True)

# Yearly mean climate
climate_yearly = (
    climate_df.groupby("year", as_index=False)[["precipitation", "temperature"]]
    .mean()
    .round(3)
)

# ============================================
# 7. PANDA POPULATION DATA (FIXED)
# ============================================
panda_points = pd.DataFrame({
    "year": [1975, 1986, 2003],
    "panda_population": [2459, 1114, 1596]
})

# include 2003 first so interpolation works properly
panda_years_full = pd.DataFrame({"year": list(range(1974, 2004))})
panda_df_full = panda_years_full.merge(panda_points, on="year", how="left")

panda_df_full["panda_population"] = panda_df_full["panda_population"].interpolate(
    method="linear",
    limit_direction="both"
)

panda_df_full["panda_population"] = panda_df_full["panda_population"].bfill()
panda_df_full["panda_population"] = panda_df_full["panda_population"].round().astype(int)

# Keep only 1974-2002
panda_df = panda_df_full[(panda_df_full["year"] >= 1974) & (panda_df_full["year"] <= 2002)].copy()

# ============================================
# 8. MERGE ALL DATA
# ============================================
merged_df = climate_yearly.merge(bamboo_yearly, on="year", how="inner")
final_df = merged_df.merge(panda_df, on="year", how="inner")
final_df = final_df.round(3)

# ============================================
# 9. EDA VISUALIZATIONS
# ============================================

# Regional precipitation
plt.figure(figsize=(10, 6))
for region_name, region_data in climate_df.groupby("region"):
    plt.plot(region_data["year"], region_data["precipitation"], marker="o", label=region_name)
plt.title("Precipitation Trend by Region (1974-2002)")
plt.xlabel("Year")
plt.ylabel("Precipitation (inches)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()

# Regional temperature
plt.figure(figsize=(10, 6))
for region_name, region_data in climate_df.groupby("region"):
    plt.plot(region_data["year"], region_data["temperature"], marker="o", label=region_name)
plt.title("Temperature Trend by Region (1974-2002)")
plt.xlabel("Year")
plt.ylabel("Temperature (°F)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()

# Bamboo occurrence
plt.figure(figsize=(10, 6))
plt.plot(bamboo_yearly["year"], bamboo_yearly["bamboo_count"], marker="o")
plt.title("Bamboo Occurrence Count by Year")
plt.xlabel("Year")
plt.ylabel("Bamboo Count")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# Bamboo index
plt.figure(figsize=(10, 6))
plt.plot(bamboo_yearly["year"], bamboo_yearly["bamboo_index"], marker="o")
plt.title("Bamboo Availability Index by Year")
plt.xlabel("Year")
plt.ylabel("Bamboo Index")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# Panda population
plt.figure(figsize=(10, 6))
plt.plot(panda_df["year"], panda_df["panda_population"], marker="o")
plt.title("Estimated Panda Population (1974-2002)")
plt.xlabel("Year")
plt.ylabel("Panda Population")
plt.ylim(0, 2600)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# Climate -> Bamboo
plt.figure(figsize=(8, 5))
plt.scatter(final_df["precipitation"], final_df["bamboo_index"])
plt.title("Precipitation vs Bamboo Availability Index")
plt.xlabel("Average Precipitation (inches)")
plt.ylabel("Bamboo Availability Index")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(final_df["temperature"], final_df["bamboo_index"])
plt.title("Temperature vs Bamboo Availability Index")
plt.xlabel("Average Temperature (°F)")
plt.ylabel("Bamboo Availability Index")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# Bamboo -> Panda
plt.figure(figsize=(8, 5))
plt.scatter(final_df["bamboo_index"], final_df["panda_population"])
plt.title("Bamboo Availability Index vs Panda Population")
plt.xlabel("Bamboo Availability Index")
plt.ylabel("Panda Population")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# ============================================
# 10. HYPOTHESIS TESTS
# ============================================
# Step 1: Climate -> Bamboo
corr_precip_bamboo, p_precip_bamboo = pearsonr(final_df["precipitation"], final_df["bamboo_index"])
corr_temp_bamboo, p_temp_bamboo = pearsonr(final_df["temperature"], final_df["bamboo_index"])

# Step 2: Bamboo -> Panda
corr_bamboo_panda, p_bamboo_panda = pearsonr(final_df["bamboo_index"], final_df["panda_population"])

# ============================================
# 11. CLEAN OUTPUT
# ============================================
print("\n=== BAMBOO AVAILABILITY DATA ===")
print(bamboo_yearly.to_string(index=False))

print("\n=== PANDA POPULATION DATA ===")
print(panda_df.to_string(index=False))

print("\n=== CLIMATE DATA ===")
print(climate_yearly.to_string(index=False))

print("\n=== FINAL MERGED DATASET ===")
print(final_df.to_string(index=False))

print("\n=== SUMMARY STATISTICS ===")
print(final_df[["panda_population", "bamboo_index", "precipitation", "temperature"]].describe().round(3))

print("\n=== HYPOTHESIS TEST RESULTS ===")
print("\n--- Climate -> Bamboo ---")
print(f"Precipitation vs Bamboo Index -> Correlation: {corr_precip_bamboo:.3f}, p-value: {p_precip_bamboo:.3f}")
print(f"Temperature vs Bamboo Index   -> Correlation: {corr_temp_bamboo:.3f}, p-value: {p_temp_bamboo:.3f}")

print("\n--- Bamboo -> Panda ---")
print(f"Bamboo Index vs Panda Population -> Correlation: {corr_bamboo_panda:.3f}, p-value: {p_bamboo_panda:.3f}")

print("\nNote: This is a preliminary stage of the project.")
print("Additional datasets such as land-use, fragmentation, bamboo coverage, and human activity indicators")
print("will be incorporated later to build a more complete Bamboo Availability Index.")

# ============================================
# 12. SAVE OUTPUT FILES
# ============================================
final_df.to_csv("final_merged_dataset.csv", index=False)
bamboo_yearly.to_csv("bamboo_yearly_index.csv", index=False)

print("\nSaved files:")
print("- final_merged_dataset.csv")
print("- bamboo_yearly_index.csv")
```
